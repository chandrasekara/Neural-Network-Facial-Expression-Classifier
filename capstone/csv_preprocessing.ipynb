```python
import pandas as pd
```

```python
# read section of the original file

data = pd.read_csv('section_fer2013.csv')
```

```python
pixel_list = []

for i in data['pixels']:
    pixel_list.append(i.split())
    #j += 1
    
pixel_list_series = pd.Series(pixel_list)

data['pixels'] = pixel_list_series

labels = data['emotion']

features = data.drop(['Usage'], axis=1, inplace=False)

print(features.head(n=5))
```

```python
# obtain total amount of pixels
num_pixels = len(features.values[0][0])
```

```python
for pixel in range(num_pixels):
    temp_list = []
    for row in range(len(data['pixels'])):
        str_pixval = features['pixels'][row][pixel]
        int_pixval = int(str_pixval)
        temp_list.append(int_pixval)
    features['pixel' + str(pixel + 1)] = pd.Series(temp_list)
    
    if pixel%20 == 0:
        print("20 checkpoint")
    
    #delete this later
    if pixel > 100:
        break
```

```python
features_new = features.drop('pixels',axis=1,inplace=False)
features_new
```

```python
features_new.to_csv("new_structure_2.csv")
```
